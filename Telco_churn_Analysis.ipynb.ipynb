```python
#Libraries Import

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
#Dataset Load

df = pd.read_csv('/content/Telco-Customer-Churn.csv')

df.head()
```

```python
#Dataset Info

print(df.shape)

print(df.columns)

df.info()
```

```python
#Missing Values
df.isnull().sum()
```

```python
#Data Cleaning

#TotalCharges Fix

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

#Remove Missing Values

df.dropna(inplace=True)

#Check Again

df.isnull().sum()
```

```python
#Churn Distribution Graph

plt.figure(figsize=(7,5))

sns.countplot(
    x='Churn',
    data=df,
    palette='Set2'
)

plt.title('Customer Churn Distribution', fontsize=16)
plt.xlabel('Churn')
plt.ylabel('Total Customers')

plt.show()
```

```python
#Pie Chart

churn_counts = df['Churn'].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    churn_counts,
    labels=churn_counts.index,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)

plt.title('Customer Churn Percentage', fontsize=16)

plt.show()
```

```python
#Contract Type Analysis

plt.figure(figsize=(10,6))

sns.countplot(
    x='Contract',
    hue='Churn',
    data=df,
    palette='coolwarm'
)

plt.title('Contract Type vs Churn', fontsize=16)

plt.xlabel('Contract Type')
plt.ylabel('Customer Count')

plt.show()
```

```python
#Monthly Charges Histogram

plt.figure(figsize=(10,6))

sns.histplot(
    df['MonthlyCharges'],
    bins=30,
    kde=True,
    color='purple'
)

plt.title('Monthly Charges Distribution', fontsize=16)

plt.xlabel('Monthly Charges')

plt.show()
```

```python
#Internet Service Analysis

plt.figure(figsize=(9,6))

sns.countplot(
    x='InternetService',
    hue='Churn',
    data=df,
    palette='viridis'
)

plt.title('Internet Service vs Churn', fontsize=16)

plt.show()
```

```python
#Tenure Boxplot

plt.figure(figsize=(9,6))

sns.boxplot(
    x='Churn',
    y='tenure',
    data=df,
    palette='Set3'
)

plt.title('Tenure vs Churn', fontsize=16)

plt.show()
```

```python
#Payment Method Analysis

plt.figure(figsize=(12,6))

sns.countplot(
    x='PaymentMethod',
    hue='Churn',
    data=df,
    palette='magma'
)

plt.xticks(rotation=25)

plt.title('Payment Method vs Churn', fontsize=16)

plt.show()
```

```python
#Correlation Heatmap

numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(10,7))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    linewidths=1
)

plt.title('Correlation Heatmap', fontsize=16)

plt.show()
```

```python
#Save Clean Dataset

df.to_csv('Cleaned_Telco_Churn.csv', index=False)
```
